```python
from google.cloud import storage
import pandas as pd

# Initialize a client
client = storage.Client()

# Define the bucket and file
bucket_name = 'retina-analytics-data'
file_name = 'Download Data csv retina (1).csv'

# Get the bucket object
bucket = client.get_bucket(bucket_name)

# Download data as pandas dataframe
blob = bucket.blob(file_name)
blob.download_to_filename('retina_data.csv')

# Load the data into pandas dataframe
df = pd.read_csv('retina_data.csv')
df.head()
```

```python
# Renaming 'Name' to 'Full Name' and 'Age' to 'Years'
data_renamed = df.rename(columns={'what is your visual acuity':'visual_acuity', 
                                    'What difficulties in seeing do u have': 'Vision_difficulties',
                                    'level of education': 'Education_level',
                                    'peripheral vision': 'Peripheral_vision',
                                    'gender': 'Gender',
                                    'disease progress degeneration': 'Degeneration',
                                    'age': 'Age',
                                    'modern or traditional technologies':'Technologies',
                                    'when do you have issues adjusting light': 'Adjusting_Light',
                                    'can you drive': 'Driving',
                                   'Unnamed: 25': 'Occupation',
                                    'Unnamed: 26': 'Disease',
                                   })
data_renamed.head(2)
```

```python
# If the column has leading/trailing spaces, strip them:
data_renamed.columns = data_renamed.columns.str.strip()
```

```python
# Selecting specific columns to create a sub-dataset
sub_dataset = data_renamed[['need', 'Education_level', 'Age', 'Occupation', 'Disease']]
# Display the first few rows of the sub-dataset
print(sub_dataset)
```

```python
# Print unique levels for each column
for column in sub_dataset.columns:
    print(f"Levels for column '{column}':")
    print(sub_dataset[column].unique())
    print()
```

```python
#map needs to 
need_mapping = {
    'assistive device': 'Assistive Device',
    'job': 'Job',
    'work training': 'Skills Training',
    'other': 'Other',
    np.nan: 'Unknown'
}
sub_dataset.loc[:, 'need'] = sub_dataset['need'].map(need_mapping)
```
